Main issue is Handling the Categorical Data and Doing EDA on it.

```python
#seprating Date,year ,month
df['Date']=df['Date_of_Journey'].str.split('/').str[0]# Splitting the date from '0' firstoccurence of '/' 
#seprating Date,year ,month
df['Month']=df['Date_of_Journey'].str.split('/').str[1]# Splitting the date from '0' firstoccurence of '/' 
#seprating Date,year ,month
df['year']=df['Date_of_Journey'].str.split('/').str[2]# Splitting the date from '0' firstoccurence of '/' 
```

```python
df.head()
df.info()#Date,month,year is still Categorical 
```

```python
#Type Conversion
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['year']=df['year'].astype(int)
df.info()
```

```python
#Dropping colunm
df.drop('Date_of_Journey',axis = 1 , inplace = True)
```

```python
df.head(2)
```

```python
# Handling Misclaneous date in the Arrival_Time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
#Handling Arrival
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_minute']=df['Arrival_Time'].str.split(':').str[1]
df.head(2)
```

```python
#Type Conversion
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_minute']=df['Arrival_minute'].astype(int) 
```

```python
#Dropping colunm
df.drop('Arrival_Time',axis = 1 , inplace = True)
```

```python
df.head()
```

```python
df['Dep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Dep_minute']=df['Dep_Time'].str.split(':').str[1]
df.head(2)
```

```python
#Type Conversion
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_minute']=df['Dep_minute'].astype(int) 
```

```python
#Dropping colunm
df.drop('Dep_Time',axis = 1 , inplace = True)
```

```python
df['Total_Stops']= df['Total_Stops'].map({'non-stop':0 , '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4,np.nan:1 })
```

```python
df.drop('Route' , axis=1,inplace=True)
```

```python
#Removing w.r.t space and 'h'  as well
# Extract hours
df['Hours'] = df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)

# Extract minutes
df['Minutes'] = df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int)

# Combine into total duration in minutes
df['TotalMinutes'] = df['Hours'] * 60 + df['Minutes']
```

**Hot Encoding**On multiple Columns

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_df =pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
# Combine one-hot encoded columns for 'Airline' into a single column
df['Airline'] = encoded_df[['Airline_Air Asia', 'Airline_Air India', 'Airline_GoAir', 'Airline_Jet Airways']] \
    .apply(lambda row: ''.join(row.astype(str)), axis=1)

# Combine one-hot encoded columns for 'Source' into a single column
df['Source'] = encoded_df[['Source_Delhi', 'Source_Mumbai']] \
    .apply(lambda row: ''.join(row.astype(str)), axis=1)

# Combine one-hot encoded columns for 'Destination' into a single column
df['Destination'] = encoded_df[['Destination_Kolkata', 'Destination_New Delhi']] \
    .apply(lambda row: ''.join(row.astype(str)), axis=1)

# Drop the original one-hot encoded columns from the DataFrame
# df = df.drop(encoded_df.columns.tolist(), axis=1)

# Display the final DataFrame with merged one-hot encoded columns
print("\nFinal DataFrame with merged encoded columns:")
print(df)
```

```python
#Dropping the Columns 
df = df.drop(['Airline', 'Source', 'Destination'], axis=1)
```

```python
#Merging it in the Data 
df = pd.concat([df, encoded_df], axis=1)
```
