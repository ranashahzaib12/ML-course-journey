**3. Data Cleaning**

```python
data.isnull().sum()# count of null values in each colunm

#Dropping all missing values Rows
data.dropna(inplace =  True)
#Removed
```

```python
# Checking the duplicate records in the data 
data.duplicated().sum()

#Seeing the suplicate Records 
# data[data.duplicated()]

#Deleting Duplicates 
data.drop_duplicates(inplace = True)

data[data.duplicated()]
```

```python
#TypeCasting 
data.dtypes

#CHanging types
data['id'] = data['id'].astype(object)
data.dtypes

data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

```python
#Geographical Distribution
plt.figure(figsize = (10,5))
sns.scatterplot(data = df , x = 'longitude' , y = 'latitude' , hue = 'room_type')
plt.title('Geographical Distribution')
plt.show()
```

```python
corr = df[['longitude', 'latitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'beds']]

# Check for NaN and drop rows with missing values
corr = corr.dropna()

# Compute the correlation matrix
corr_matrix = corr.corr()

# Create the heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(data=corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
