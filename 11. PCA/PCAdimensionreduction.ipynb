- Applying PCA Algorithim 

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(n_components=2) # How many Features We want out of ALL 
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca
# Reduced to two columns Only 
# Purpose is to Visulaize it
```

```python
# The Amount of Variance we are Able to get 
pca.explained_variance_
```
